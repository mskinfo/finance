```python
import pandas as pd
```

```python
vixdata = pd.read_excel(r"/content/vix_data.xlsx")
vixdata
```

```python
from sklearn import linear_model
X = vixdata[['T1OPEN', 'T1HIGH', 'T1LOW', 'T1CLOSE', 'T2OPEN', 'T2HIGH', 'T2LOW', 'T2CLOSE', 'T3OPEN', 'T3HIGH', 'T3LOW', 'T3CLOSE', 'T4OPEN', 'T4HIGH', 'T4LOW', 'T4CLOSE']]
y = vixdata['CLOSE']
```

```python
train_x = X[7000:8000]
train_x
```

```python
train_y = y[7000:8000]
train_y
```

```python
test_x = X[8000:]
test_x
```

```python
test_y = y[8000:]
test_y
```

```python
regr = linear_model.LinearRegression()
regr.fit(train_x, train_y)

predictions = regr.predict(test_x)
preds = pd.DataFrame(predictions)
test_y = pd.DataFrame(test_y)

test_y.to_excel("testout.xlsx")
preds.to_excel("predsout.xlsx")
```

# **Theory**
This round of analysis used 1000 days of historical cboe vix data to create a model, which, using historical inputs, attempts to predict the change between the previous close and the next day's close for 288 days.

The model predicts the price direction with an accuracy of 52.621% in terms of the binary direction.

If an agent purchased or shorted the vix based on the predicted binary direction, the average return per day is 0.21%.  Annually, this is 68.95%.

Due to the wash-sale rule, implementing this strategy would be a flawed decision.  The wash-sale rule states that if an agent closes a trade at a loss and opens the same or a substantially identical trade within 30 calendar days before or after the first trade, they will not be able to take a loss for that security on your their tax return.

To avoid this rule, the agent could use futures, or options on futures for the underlying security, as the wash-sale rule does not apply to these instruments.

# **Implementation**
Commissions on options are the only variable cost, at USD 0.65 per trade.  There will be two trades per day, making the monthly trading costs equivalent to $26.  From here, options will magnify returns by a factor of 10.  So, each day, the strategy will yield 0.21% * 10 = 2.1%.

Within the first year, starting with USD 500 risked, this will yield USD 90,241.
